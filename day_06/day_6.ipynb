```python
# 87.135315,27.050159,89.224091,27.906655
# Bottom Left Long, Bottom Left Lat, Top Right Long, Top Right Lat)
kan_box = 87.135315,27.050159,89.224091,27.906655
```

```python
rm = RidgeMap(kan_box)
values = rm.get_elevation_data(num_lines=150)

values=rm.preprocess(
    values=values,
    lake_flatness=2,
    water_ntile=10,
    vertical_ratio=240)
rm.plot_map(values=values,
            label='Kanchenjunga, INDIA',
            label_y=0.1,
            label_x=0.55,
            label_size=40,
            linewidth=2,
            line_color = plt.get_cmap('spring'))
```
